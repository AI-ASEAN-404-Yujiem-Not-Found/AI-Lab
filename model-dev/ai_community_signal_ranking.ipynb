Simulate Opportunity Diffusion

```python
import numpy as np

propagation_probability = 0.3

active_nodes = set(top_receivers['UserID'].tolist())

activated_nodes_over_time = []

iteration = 0
max_iterations = 20

print(f"Simulation started with {len(active_nodes)} initial active nodes.")

while True and iteration < max_iterations:
    iteration += 1
    newly_activated_nodes = set()
    current_step_activations = 0

    for node in list(active_nodes):
        for neighbor in G.neighbors(node):
            if neighbor not in active_nodes and neighbor not in newly_activated_nodes:
                if np.random.rand() <= propagation_probability:
                    newly_activated_nodes.add(neighbor)
                    current_step_activations += 1

    activated_nodes_over_time.append(current_step_activations)

    if not newly_activated_nodes:
        print(f"No new nodes activated in iteration {iteration}. Simulation stopping.")
        break

    active_nodes.update(newly_activated_nodes)
    print(f"Iteration {iteration}: {current_step_activations} new nodes activated. Total active: {len(active_nodes)}")

print(f"\nTotal unique nodes activated after simulation: {len(active_nodes)}")
print(f"Newly activated nodes per iteration: {activated_nodes_over_time}")
```

```python
import torch
from torch_geometric.data import Data
import pandas as pd
import torch.nn as nn
from torch_geometric.nn import SAGEConv
import torch.optim as optim
import numpy as np
from community import community_louvain
import networkx as nx
import gdown

num_users = 200
num_villages = 5
occupations = ['Farmer', 'Trader', 'Craftsman', 'Teacher', 'Doctor', 'Artisan']

users_df = pd.DataFrame({
    "UserID": [f"U{i+1}" for i in range(num_users)],
    "village_id": np.random.randint(1, num_villages + 1, num_users),
    "occupation": np.random.choice(occupations, num_users),
    "participation_rate": np.round(np.random.rand(num_users), 2),
    "microtask_completion": np.round(np.random.rand(num_users), 2),
    "trust_score": np.round(np.random.rand(num_users), 2)
})

num_edges = 500
source_users = np.random.choice(users_df['UserID'], num_edges)
target_users = np.random.choice(users_df['UserID'], num_edges)

edges_list = []
for s, t in zip(source_users, target_users):
    if s != t:
        edge = tuple(sorted((s, t)))
        if edge not in [tuple(sorted((x['source_user'], x['target_user']))) for x in edges_list]:
            edges_list.append({
                "source_user": s,
                "target_user": t,
                "interaction_weight": np.round(np.random.rand(), 2)
            })

edges_df = pd.DataFrame(edges_list)

nodes_url_2 = 'https://drive.google.com/file/d/1T0wMgujwsAk4sqpFSGNP7-e6W3Ou2bPH/view?usp=drive_link'
file_id_nodes_2 = nodes_url_2.split('/')[-2]
gdown.download(id=file_id_nodes_2, output='community_nodes_new.csv', quiet=True)
community_nodes_df = pd.read_csv('community_nodes_new.csv')

edges_url_2 = 'https://drive.google.com/file/d/1pXi23RHrkngjYGo5bjtsOwYdQkrRcQfM/view?usp=drive_link'
file_id_edges_2 = edges_url_2.split('/')[-2]
gdown.download(id=file_id_edges_2, output='community_edges_new.csv', quiet=True)
community_edges_df = pd.read_csv('community_edges_new.csv')

gdp = 'https://drive.google.com/file/d/13oOu6M74yw5k3mYYsfjTsg4DnGJ9aLr9/view?usp=drive_link'
file_id_gdp = gdp.split('/')[-2]
gdown.download(id=file_id_gdp, output='messages_new.csv', quiet=True)
messages_df = pd.read_csv('messages_new.csv')

community_nodes_df['obs_value'] = community_nodes_df['obs_value'].fillna(community_nodes_df['obs_value'].mean())
messages_df['obs_value'] = messages_df['obs_value'].fillna(messages_df['obs_value'].mean())

community_nodes_df['time'] = community_nodes_df['time'].str[:4].astype(int)
community_edges_df['time'] = community_edges_df['time'].str[:4].astype(int)

rename_mapping = {
    'ref_area.label': 'country',
    'source.label': 'data_source',
    'indicator.label': 'indicator'
}
community_nodes_df = community_nodes_df.rename(columns=rename_mapping)
community_edges_df = community_edges_df.rename(columns=rename_mapping)
messages_df = messages_df.rename(columns=rename_mapping)

village_country_mapping = {
    1: 'Bolivia (Plurinational State of)',
    2: 'Australia',
    3: 'Canada',
    4: 'Chile',
    5: 'Colombia'
}
users_df['country'] = users_df['village_id'].map(village_country_mapping)

aggregated_nodes_df = community_nodes_df.groupby('country')['obs_value'].mean().reset_index()
aggregated_nodes_df = aggregated_nodes_df.rename(columns={'obs_value': 'avg_nodes_obs_value'})

aggregated_edges_df = community_edges_df.groupby('country')['obs_value'].mean().reset_index()
aggregated_edges_df = aggregated_edges_df.rename(columns={'obs_value': 'avg_edges_obs_value'})

aggregated_messages_df = messages_df.groupby('country')['obs_value'].mean().reset_index()
aggregated_messages_df = aggregated_messages_df.rename(columns={'obs_value': 'avg_messages_obs_value'})

country_features_df = pd.merge(aggregated_nodes_df, aggregated_edges_df, on='country', how='outer')
country_features_df = pd.merge(country_features_df, aggregated_messages_df, on='country', how='outer')
country_features_df = country_features_df.fillna(0)

users_df = pd.merge(users_df, country_features_df, on='country', how='left')

G = nx.Graph()

for index, row in users_df.iterrows():
    user_id = row['UserID']
    attributes = row.drop(['UserID', 'village_id', 'country']).to_dict() # 'village_id' and 'country' are merged as features now
    G.add_node(user_id, **attributes)

for index, row in edges_df.iterrows():
    source = row['source_user']
    target = row['target_user']
    weight = row['interaction_weight']
    G.add_edge(source, target, interaction_weight=weight)

degree_centrality = nx.degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)
pagerank = nx.pagerank(G)

for user_id in G.nodes():
    G.nodes[user_id]['degree_centrality'] = degree_centrality.get(user_id, 0.0)
    G.nodes[user_id]['betweenness_centrality'] = betweenness_centrality.get(user_id, 0.0)
    G.nodes[user_id]['pagerank'] = pagerank.get(user_id, 0.0)


node_data = []
for user_id, attributes in G.nodes(data=True):
    node_features = {'UserID': user_id}
    node_features.update(attributes)
    node_data.append(node_features)

node_features_df = pd.DataFrame(node_data)

node_features_df['community_role_score'] = (
    0.4 * node_features_df['pagerank'] +
    0.2 * node_features_df['degree_centrality'] +
    0.2 * node_features_df['trust_score'] +
    0.2 * node_features_df['participation_rate']
)

partition = community_louvain.best_partition(G)
community_df = pd.DataFrame(list(partition.items()), columns=['UserID', 'community_id'])
node_features_df = pd.merge(node_features_df, community_df, on='UserID', how='left')

if 'occupation' in node_features_df.columns:
    node_features_df = pd.get_dummies(node_features_df, columns=['occupation'], prefix='occupation')
else:
    print("Warning: 'occupation' column not found, assuming it has already been one-hot encoded.")

numerical_features_for_gnn = [
    'participation_rate',
    'microtask_completion',
    'trust_score',
    'avg_nodes_obs_value',
    'avg_edges_obs_value',
    'avg_messages_obs_value',
    'degree_centrality',
    'betweenness_centrality',
    'pagerank',
    'community_role_score'
]

occupation_cols = [col for col in node_features_df.columns if col.startswith('occupation_')]
numerical_features_for_gnn.extend(occupation_cols)

existing_features = [feature for feature in numerical_features_for_gnn if feature in node_features_df.columns]

for col in existing_features:
    if node_features_df[col].dtype == 'bool':
        node_features_df[col] = node_features_df[col].astype(int)
    else:
        node_features_df[col] = pd.to_numeric(node_features_df[col], errors='coerce')
node_features_df[existing_features] = node_features_df[existing_features].fillna(0)

print("Dtypes of features selected for GNN input 'x':")
print(node_features_df[existing_features].dtypes)

x = torch.tensor(node_features_df[existing_features].values, dtype=torch.float)

user_id_to_idx = {user_id: idx for idx, user_id in enumerate(node_features_df['UserID'])}

remapped_edge_index = []
for u, v in G.edges():
    if u in user_id_to_idx and v in user_id_to_idx:
        remapped_edge_index.append([user_id_to_idx[u], user_id_to_idx[v]])
    else:
        print(f"Warning: Edge ({u}, {v}) contains node(s) not in processed node_features_df. Skipping edge.")

edge_index = torch.tensor(remapped_edge_index, dtype=torch.long).t().contiguous()

data = Data(x=x, edge_index=edge_index)

print("PyTorch Geometric Data object created.")
print(f"Node features (x) shape: {data.x.shape}")
print(f"Edge index (edge_index) shape: {data.edge_index.shape}")

class GraphSAGE(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GraphSAGE, self).__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, out_channels)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(p=0.5)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = self.relu(x)
        x = self.dropout(x)
        x = self.conv2(x, edge_index)
        return x

in_channels = data.num_node_features
hidden_channels = 64
out_channels = 1

model = GraphSAGE(in_channels, hidden_channels, out_channels)

print("\nGraphSAGE Model Architecture (redefined for scalar output):")
print(model)

optimizer = optim.Adam(model.parameters(), lr=0.01)
criterion = nn.MSELoss()

y_target = torch.tensor(node_features_df['community_role_score'].values, dtype=torch.float).view(-1, 1)

num_epochs = 50

print("\n--- Starting Conceptual Training Loop ---")
model.train()
for epoch in range(num_epochs):
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = criterion(out, y_target)
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

print("--- Conceptual Training Loop Finished ---")

print("\n--- Conceptual Inference Process ---")
model.eval()
with torch.no_grad():
    gnn_influence_scores = model(data.x, data.edge_index)

print(f"Inference complete. GNN-derived influence scores shape: {gnn_influence_scores.shape}")
print("These scores represent the learned representations or predicted community_role_scores for each node based on the GNN model.")
print("First 5 GNN-derived influence scores:\n", gnn_influence_scores[:5].squeeze().tolist())

node_features_df['gnn_influence_score'] = gnn_influence_scores.cpu().numpy().flatten()
print("\nNode Features DataFrame with GNN Influence Score Head:")
print(node_features_df.head())
```
